```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv("HousingData.csv")
```

```python
data
```

```python
data.info()
```

```python
data.drop(columns=['NOX', 'PTRATIO', 'B'], inplace=True)
```

```python
data
```

```python
data.info()
```

```python
# CRIM - MEAN
# ZN - 0.0
# INDUS - mean
# CHAS - 0
# AGE - mean
# LSTAT - mean
```

```python
data['CRIM'].fillna(data['CRIM'].mean(), inplace=True)
data['ZN'].fillna(0.0, inplace=True)
data['INDUS'].fillna(data['INDUS'].mean(), inplace=True)
data['CHAS'].fillna(0, inplace=True)
data['AGE'].fillna(data['AGE'].mean(), inplace=True)
data['LSTAT'].fillna(data['LSTAT'].mean(), inplace=True)
```

```python
data.info()
```

```python
X = data.drop(['MEDV'], axis=1)
y = data['MEDV']
```

```python
# train test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
# data exploration
```

```python
train_data = X_train.join(y_train)
```

```python
train_data.hist(figsize=(15, 8))
```

```python
# Standardization 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
# model
from sklearn.linear_model import LinearRegression
# cross validation
from sklearn.model_selection import cross_val_score
```

```python
regression = LinearRegression()
regression.fit(X_train, y_train)
```

```python
mse = cross_val_score(regression, X_train, y_train, scoring='neg_mean_squared_error', cv=10)
```

```python
np.mean(mse)
```

```python
# prediction
reg_pred = regression.predict(X_test)
```

```python
reg_pred
```

```python
import seaborn as sns
sns.displot(reg_pred-y_test, kind='kde')
```

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(reg_pred,y_test)
score
```
